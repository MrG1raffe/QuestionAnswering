## Duckduckgo avec Langchain

Voir <https://python.langchain.com/docs/integrations/tools/ddg>

```python
from DuckDuckGo_search import duckduckgo_search
from mdeberta_request import mdeberta_request

n_batch = 1000
for i in range(n_batch):
    question, answer = dataset[i]['question'], dataset[i]['answer']
    search_result = duckduckgo_search(question)
    answer_model = mdeberta_request(question=question, context=search_result)
    print('\n\nQuestion:', question)
    print('\nSearch result:')
    print(search_result)
    print('\nModel answer:', answer_model)
    print('Answer (ground truth):', answer)
    print('\n')
```

```python
from transformers import pipeline

pipe = pipeline("text-generation", model="gpt2")
pipe("what is a lobster roll")
```
